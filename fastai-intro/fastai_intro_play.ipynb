````python
# Path('/root/.fastai/data/oxford-iiit-pet/images/english_setter_32.jpg') # Dog
# Path('/root/.fastai/data/oxford-iiit-pet/images/Sphynx_146.jpg' # Cat
a_pet_path = Path("/root/.fastai/data/oxford-iiit-pet/images/Sphynx_146.jpg")
a_pet_path.stem
````

````python
def is_cat(x): return x.stem[0].isupper()

is_cat(a_pet_path)
````

````python
# Look for images recursively in a root folder
imgs = get_image_files(path); imgs
````

## Splitting the data
- into 'train' and 'valid' (validation) splits - 80%-20% split

````python
splitter = RandomSplitter(valid_pct=0.2, seed=42); splitter
display(splitter)
print(splitter(imgs[:10]))
print(splitter(imgs[:10]))
````

````python
# model_blueprint_and_weights = mycool_ml_lib.train(
#                     input_type="image", <------
#                     output_type="category", <------------
#                     imgs,
#                     ..
#                 )

def add(a: int, b: int):
  return a + b

display(add)
display(ImageBlock)
display(CategoryBlock)
````

````python
# DataBlock. Data Structure and Transformation Specification

# model_blueprint_and_weights = mycool_ml_lib.train(
#                     input_type="image",
#                     output_type="category",
#                     imgs,
#                     labels, # for each image i.e cat or dog
#                     splitter,
#                     model_blueprint, <----------- Defined Later

pets_dblock = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    get_y=is_cat,
    splitter=splitter,
    item_tfms=Resize(224))
````

````python
# DataLoader. e.g. It provides batches of a few items at a time to the GPU.
pet_dataloader = pets_dblock.dataloaders(path)
````

## Metric
* Metric: Assess the model's 'accuracy' using the validation set. 
* Maps to business/task criteria. For Humans

## Loss Function
* Loss Function. Assess the model's 'accuracy' AND
    * **give hints on how to get closer to the optimal solution. smart 'trial and error'**   
* 'Hot and Cold' game

````python
print("Metric")
display(error_rate) #  i.e. how many did we get wrong. opposite of accuracy
print("Loss Function")
cross_entropy_func = CrossEntropyLossFlat()
display(cross_entropy_func)
````

````python
learn = vision_learner(pet_dataloader,
                       resnet18, 
                       metrics=error_rate, 
                       loss_func=cross_entropy_func # <-- loss_func is optional.
                      )
````

# Train vs Fine-Tuning
```python
# model_blueprint_and_weights = mycool_ml_lib.train(...)
```
* Train is to train a model from scratch
* Also called `fit`. i.e. fit a `model architecture` with weights to our particular task
* But our model is already pretrained. We need to `fine-tune` it to our model.

````python

learn.fine_tune(1)
````

## ENOUGH! 5 lines of code please!

````python
#id first_training
#caption Results from the first training

from fastai.vision.all import *
path = untar_data(URLs.PETS)/'images'

# ~5 lines of code!
def is_cat(x): return x[0].isupper()
dls = ImageDataLoaders.from_name_func(path, get_image_files(path), valid_pct=0.2, seed=42, label_func=is_cat, item_tfms=Resize(224))
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(1)
````

````python
test_img = imgs[-1] # JUST FOR PRESENTATION!
img = PILImage.create(test_img)
img.to_thumb(192)
````

````python
img = test_img
is_cat, category,probs = learn.predict(img)

print(is_cat)
print(category)
print(probs)
print(probs[category])
````
